```python
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from matplotlib import interactive
```

### KNN for mean data 

```python
data = pandas.read_excel("mean_data_final.xlsx")
cl=data.columns
X=data[cl[1:len(cl)-1]]
y=data['Rings']
```

```python
from sklearn.model_selection import train_test_split
# dividing X, y into train and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y,  train_size=0.7, test_size=0.3, random_state=123) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
neighbors = list(range(1,30))
train_results = []
test_results = []
scores=[]
for n in neighbors:
    model = KNeighborsClassifier(n_neighbors=n)
    model.fit(X_train, y_train)
    train_pred = model.predict(X_train)
    accuracy = model.score(X_test, y_test) 
    print('Accuracy for k = '+str(n),accuracy) 
    scores.append(accuracy)
# import Matplotlib (scientific plotting library)
import matplotlib.pyplot as plt

# allow plots to appear within the notebook
%matplotlib inline

# plot the relationship between K and testing accuracy
# plt.plot(x_axis, y_axis)
plt.plot(neighbors, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
  

   
```

```python
distances = [1, 2, 3, 4, 5]
train_results = []
test_results = []
scores=[]
for p in distances:
    model = KNeighborsClassifier(p=p)
    model.fit(X_train, y_train)
    train_pred = model.predict(X_train)
    accuracy = model.score(X_test, y_test) 
    print('Accuracy for p = '+str(p),accuracy)
    scores.append(accuracy)
# import Matplotlib (scientific plotting library)
import matplotlib.pyplot as plt

# allow plots to appear within the notebook
%matplotlib inline

# plot the relationship between K and testing accuracy
# plt.plot(x_axis, y_axis)
plt.plot(distances, scores)
plt.xlabel('Value of p for KNN')
plt.ylabel('Testing Accuracy')
  
   
```
